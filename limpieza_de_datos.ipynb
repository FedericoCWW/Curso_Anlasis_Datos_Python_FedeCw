```python
import pandas as pd
```

```python
data = {"id": [1001,1002,1003,1003],
       "cant": [30, None, 25, 200],
       "precio": [20.6, None, None, 43.6]}
df = pd.DataFrame(data)
```

```python
df
```

```python
df_eliminados = df.dropna()
```

```python
df.isnull().sum()
```

```python
df_eliminados
```

```python
valores = {"cant": 0, "precio": df["precio"].mean()}
df_ceros = df.fillna(valores)
```

```python
df_ceros
```

```python
df_ceros["cant"] = df_ceros["cant"].astype(int)
df_ceros
```

```python
df_unicos = df_ceros.drop_duplicates(subset="id")
df_unicos
```
