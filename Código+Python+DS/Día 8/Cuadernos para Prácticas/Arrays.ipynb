```python
lista_grande = list(range(100000000))
```

```python
arr_grande = np.array(lista_grande)
```

```python
import time
```

```python
# Verificacion tiempo lista
init = time.time()
for x in lista_grande:
    x ** 2
end = time.time()
print(f"Tiempo Lista" , end - init)
```

```python
# Verificacion tiempo arr Numpy
init = time.time()
arr_grande ** 2
end = time.time()
print(f"Tiempo Array Numpy" , end - init)
```
