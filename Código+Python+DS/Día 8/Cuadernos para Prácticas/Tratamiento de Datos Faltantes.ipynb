```python
import numpy as np
```

```python
arr = np.array([1,2,3, np.nan, 5])
```

```python
arr
```

```python
np.isnan(arr)
```

```python
np.nanmean(arr)
```

```python
arr2 = np.where(np.isnan(arr), 0, arr)
arr2
```

```python
prom = np.nanmean(arr)
arr2 = np.where(np.isnan(arr), np.ceil(prom), arr)
arr2
```

```python
arr_filtr = arr[~np.isnan(arr)]
arr_filtr
```
