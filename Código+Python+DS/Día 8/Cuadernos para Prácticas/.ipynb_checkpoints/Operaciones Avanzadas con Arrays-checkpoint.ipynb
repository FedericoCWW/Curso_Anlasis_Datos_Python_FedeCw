```python
import numpy as np
```

```python
array1 = np.array([1, 2, 3])
array1
```

```python
array2 = np.array([[0], [10], [20], [30]])
array2
```

```python
# broadcasting: describes how NumPy treats arrays with different shapes during arithmetic operations.
suma_arr = array2 + array1
suma_arr
```

```python
suma_mult = array2 * array1
suma_mult
```

```python
# funciones universales (ufuncs)
a = np.array([7,2,1])
b = np.array([7,6,3])
```

```python
result = np.add(a,b)
```
