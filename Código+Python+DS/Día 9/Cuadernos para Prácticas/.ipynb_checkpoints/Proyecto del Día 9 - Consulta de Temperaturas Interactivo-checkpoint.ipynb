```python
import pandas as pd
import matplotlib.pyplot as plt

import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\Usuario\Documents\COSAS PYTHON DATA SCIENCE Y ML UDEMY)\Datos+Meteorológicos_Arg_2023.csv")
df
```

```python
df['Fecha'] = pd.to_datetime(df['Fecha'], format='%d/%m/%Y')
print(df['Fecha'].dtype)  # Ahora será datetime64[ns]
```

```python
df
```

```python
# esta es la consulta que necesito para realizar el grafico
df_select = df[(df['Ciudad'] == 'Rosario') & (df['Fecha'].dt.month == 2)]
df_select
```

```python
def Consulta_Temperatura_Interactivo():
    seguir = True
    while seguir:
        select_ciudad = input("Por favor seleccione una ciudad")
        select_mes = int(input("Por favor seleccione un mes (en numero)"))
        
        df_filtrado = df[(df['Ciudad'] == select_ciudad) & (df['Fecha'].dt.month == select_mes)] 
        
        x = np.arange(len(df_filtrado))
        width = 0.35
    
        fig, ax = plt.subplots(figsize=(12, 6), layout='constrained')
        bar_min = ax.bar(x,df_filtrado['Temperatura Maxima'], label='Temp Maxima', color='red', alpha=0.8)
        bar_max = ax.bar(x + width/2,df_filtrado['Temperatura Minima'], label='Temp Minima', color='blue', alpha=0.8)
        
        ax.set_xlabel('Días')
        ax.set_ylabel('Temperatura (°C)')
        ax.set_title(f'Temperaturas Mínimas y Máximas - {select_ciudad}')
        ax.set_xticks(x)
        ax.set_xticklabels(df_filtrado['Fecha'].dt.strftime('%d/%m'), rotation=45)
        ax.legend()
        ax.grid(axis='y', alpha=0.3)

        plt.show()
        
        select_seguir = int(input("¿Desea seguir usando el programa? (1 para si, 2 para no)"))
        if select_seguir == 2:
            seguir = False
        
    
```

```python
Consulta_Temperatura_Interactivo()
```
