```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\Usuario\Documents\COSAS PYTHON DATA SCIENCE Y ML UDEMY)\Iluminacion.csv")
df
```

```python
fig, ax = plt.subplots()

hb = ax.hexbin(df['Latitud'], df['Longitud'], C=df['Luminosidad'], gridsize=40, cmap='Blues')
plt.colorbar(hb,label='Nivel de sombras')
plt.xlabel("Latitud")
plt.ylabel("Longitud")
plt.title("Estadistica luminocidad");
```
