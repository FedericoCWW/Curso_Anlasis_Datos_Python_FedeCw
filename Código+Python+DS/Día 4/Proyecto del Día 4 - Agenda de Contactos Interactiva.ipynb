## Resolución

Crea una agenda vacía

```python
contactos = {}
```

En la siguiente celda, incluye el código para ejecutar las siguientes instrucciones:
- Mostrar en pantalla las opciones
- Pedirle al usuario que elija las opciones
- Crear una estructura de control (*if*, *elif* y *else*) que ejecute diferentes bloques de código según las opciones elegidas.

```python
print("Elje una opcion. 1: agregar contacto; 2: buscar contacto; 3: editar contacto:; 4: eliminar contacto; 5: mostrar todos los contacto")
eleccion = int(input())
if 0 < eleccion <= 5:
    if eleccion == 1:
        nombre = input("Escribe nombre de contacto: ")
        tel = input("Escribe el telefono del contacto:")
        contactos[nombre] = tel
    elif eleccion == 2:
        nombre = input("Ingrese Nombre del contacto a buscar: ")
        if nombre in contactos:
            print(contactos[nombre])
        else:
            print("contacto no existente!")
    elif eleccion == 3:
        nombre = input("Ingrese Nombre del contacto a editar: ")
        if nombre in contactos:
            nuevo_tel = input("Escribe el telefono del contacto a editar:")
            contactos[nombre] = nuevo_tel
            print(contactos[nombre])
        else:
            print("contacto no existente!")
    elif eleccion == 4:
        nombre = input("Ingrese Nombre del contacto a eliminar: ")
        if nombre in contactos:
            contactos.pop(nombre)
            print("Se elimino el contacto")
        else:
            print("contacto no existente!")
    else:
        print(contactos)
else:
    print("la eleccion que eligio no corresponde a una de las brindadas")
```
